```python
import os
import shutil
```

```python
sourceDataset = "../Dataset/"

path_robotDataset = sourceDataset + "Robots Datasets/"
path_kaggleDataset = sourceDataset + "Merged Dataset Kaggle/"
path_allDataset = sourceDataset + "Merged Dataset All/"
```

```python
robotDict = {}
kaggleDict = {}

for folder in os.listdir(path_robotDataset):
    robotDict[folder] = len(os.listdir(path_robotDataset + folder))

for folder in os.listdir(path_kaggleDataset):
    kaggleDict[folder] = len(os.listdir(path_kaggleDataset + folder))
```

```python
temp = []
sameClass = []

print(f"Dataset Robot ==> {len(robotDict)} class\n")
for key, value in robotDict.items():
    print(f"class {key} have {value} images")
    temp.append(key)
    
print(f"\n === \nDataset Kaggle ==> {len(kaggleDict)} class\n")
for key, value in kaggleDict.items():
    print(f"class {key} have {value} images")
    
    try:
        sameClass.append(temp[temp.index(key)])
    except:
        pass
```

```python
print("Class that belongs to two dataset is:")
for index, value in enumerate(sameClass):
    print(f"{index+1}. {value}")
```

```python
robot_class_list = [value for value in robotDict]
kaggle_class_list = [value for value in kaggleDict]
```

```python
def copyFiles(classList, sourcePath, destPath):
    if os.listdir(destPath):
        for folder in os.listdir(destPath):
            try:
                label_index = classList.index(folder)
                for file in os.listdir(sourcePath + classList[label_index]):
                    print(file)
                    shutil.copy(sourcePath + classList[label_index] + "/" + file, destPath + folder)
            except:
                print("error")
    else:
        for folder in classList:
            shutil.copytree(sourcePath + folder, destPath + folder)
```

```python
copyFiles(kaggle_class_list, path_kaggleDataset, path_allDataset)
#JANGAN di RUN LAGI
```

```python
copyFiles(robot_class_list, path_robotDataset, path_allDataset)
```
